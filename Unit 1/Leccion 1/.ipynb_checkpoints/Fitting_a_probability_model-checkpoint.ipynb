# Fitting a Probability Model

In this document we will discuss how to use (well-known) probability distributions to model univariate data (a single variable) in Python. We will call this process "fitting" a model.

For the purpose of this document, the variables that we would like to model are assumed to be a random sample from some population. That is, we are considering models of the form
$$
X_1,X_2,\ldots,X_n \overset{idd}{\sim} f(x,\theta)
$$

Models of this form assume that the data we obtained can be modeled by random variables that are independent and identically distributed according to some distribution with parameter $\theta$. It is important to be aware that these are assumptions of the model that we are using. For now, we will use these (often implicit) assumptions, but not necessarily spend much time challenging them. (Specifically, we won't criticize the independence and identical assumptions. If our data truly is a random sample, there shouldn't be much of an issue.) For other more explicit assumptions (the distribution assumption) that we make, we will learn how to criticize them.

Some natural questions should arise when modeling a variable:

- What probability models are appropriate? Specifically, what distribution should be used?
- How do you fit a particular model? That is, how do we fit a chosen probability distribution to some data?
- How do we know if a model that we fit is good?

To fit a probability model and answer these questions, we will generally use the following procedure:

- Look at the data.
- Assume a probability distribution.
- Estimate the model parameters, that is, the parameters of the chosen distribution.
- Check how well the estimated model fits using a QQ-Plot.
- To illustrate this process and introduce some new concepts such as a QQ-Plot, let’s look a several examples.

## Hours of Sleep, Normal Model

Let's create a new variable called `hour_sleep` that measures the hours of sleep each night that we will attempt to model.

```python
hour_sleep = fitbit_sleep.min_sleep / 60

plt.hist(hour_sleep, range=(4,12), density=True, bins=16)
plt.title('Histogram of hours of sleep')
plt.xlabel('Hours of sleep')
plt.ylabel('Density')
plt.grid()
```

This histogram gives some insight into why we wouldn't simply use the empirical distribution. In some sense, we can think of using the empirical distribution as “overfitting.” It places zero probability on too much of the range of the variable, and places too much probability on the observations we have observed. By fitting a probability model, we will "smooth" away these issues.

Here, we can argue that a continuous distribution would make the most sense. While the Fitbit only records sleep to the nearest minute, in reality sleep time could take on any positive real number between 0 and 24 hours.

Without giving it too much though, we'll attempt to model this data with a normal distribution. While a normal distribution does allow for the possibility of any real number, the vast majority of the probability is within three standard deviation of the mean.

To "fit" the normal model, we simply need to estimate its two parameters, $\mu$ and $\sigma^2$. Recall that the probability density function for a normal random variable is given by
$$
f(x|\mu,\sigma^2) = \frac{1}{\sigma\sqrt{2\pi}}\exp\left[  -\left( \frac{x-\mu}{\sigma}  \right)^2 \right],
\quad x\in\mathbb{R},\ \mu\in\mathbb{R},\ \sigma^2>0
$$

The maximum likelihood estimators for $\mu$ and $\sigma^2$ are given by
$$
\hat\mu = \bar{x} = \frac{1}{n}\sum_{i=1}^n x_i\\
\hat\sigma^2 = \frac{1}{n}\sum_{i=1}^n (x_i-\bar{x})^2
$$

Scipy already implements the MLE for all included distributions into the fit method.

```python
from scipy import stats
```

```python
mu, sigma = stats.norm.fit(hour_sleep)
print(f'mu={mu:.2f}, sigma={sigma:.2f}')
```

Here we see these two estimates. In this dataset, Dave sleeps on average 7.4 hours, with a standard deviation of about 0.8 hours.

To get a sense of how well this model fits, we add its density to the histogram of the data. Note that the `norm.pdf()` function takes the standard deviation, not the variance as an argument.

```python
plt.hist(hour_sleep, range=(4,12), density=True, bins=16)
plt.title('Histogram of hours of sleep')
plt.xlabel('Hours of sleep')
plt.ylabel('Density')
plt.grid()

mu, sigma = stats.norm.fit(hour_sleep)

grid = np.linspace(4,12,100)
plt.plot(grid, stats.norm.pdf(grid, mu, sigma), label=f'$\hat\mu={mu:.2f}$, $\sigma^2={sigma**2:.2f}$')
plt.legend()
```

It turns out that it's difficult to tell how well this fits using only a histogram. It actually becomes really difficult when you realize that histograms are very sensitive to bin width selection.

```python
fig, ax = plt.subplots(1,2, figsize=(12,4))

ax[0].hist(hour_sleep, range=(4,12), density=True, bins=16)
ax[0].set_title('Histogram of hours of sleep')
ax[0].set_xlabel('Hours of sleep')
ax[0].set_ylabel('Density')
ax[0].grid()
mu, sigma = stats.norm.fit(hour_sleep)
grid = np.linspace(4,12,100)
ax[0].plot(grid, stats.norm.pdf(grid, mu, sigma), label=f'$\hat\mu={mu:.2f}$, $\sigma^2={sigma**2:.2f}$')

ax[1].hist(hour_sleep, range=(4,12), density=True, bins=40)
ax[1].set_title('Histogram of hours of sleep')
ax[1].set_xlabel('Hours of sleep')
ax[1].set_ylabel('Density')
ax[1].grid()
ax[1].plot(grid, stats.norm.pdf(grid, mu, sigma), label=f'$\hat\mu={mu:.2f}$, $\sigma^2={sigma**2:.2f}$')
```

Instead of using a histogram to check how well our model fits, we'll instead use a quantile-quantile plot, often simply called a QQ plot. We’ll save the details for later, but the idea of a QQ plot is simple. One one axis, normally the y axis, we plot the actual data that we observed. On the other axis, the x axis, we plot where we would expect the data to be if it was sampled from the distribution that we fit. We then add a line for $y=x$. If the model fits well, the points should fall close to the line. If the model fits poorly, the points will deviate form the line.

```python
fig, ax = plt.subplots()
res = stats.probplot(hour_sleep, sparams=(mu, sigma), dist=stats.norm, plot=ax)
ax.set(title='QQ-Plot: Hours of Sleep, Normal Distribution', 
       xlabel='Theoretical Quantiles, Normal Distribution',
       ylabel='Sample Quantiles, Hours of Sleep');
```

## Awakenings, Poisson Model

Now let’s try to model the number of times Dave wakes up during the night.

```python
awakes = fitbit_sleep.num_awake
```

```python
np.histogram(awakes)
```

```python
bars = np.unique(awakes, return_counts=True)
plt.bar(bars[0], bars[1]/len(awakes))
plt.ylim(0, 0.15)
plt.xlim(0,55)
plt.grid()
plt.title('Fitbit Sleep Data: Number of Awakenings')
plt.xlabel('Number of Awakenings')
```

Let's try a Poisson model. To "fit" the model, we need to estimate λ, which we have repeatedly seen can be done using the the sample mean, $\bar{x}$. (Scipy doesn't include a fit method for discrete distributions).

```python
l = np.mean(awakes)
print(l)
```

When we represent this distribution as a barplot, we obtain the following.

```python
grid = np.arange(55)
plt.bar(grid, stats.poisson.pmf(grid, l))
plt.title('Fitbit Sleep Data: Number of Awakenings')
plt.xlabel('Number of Awakenings')
plt.ylim(0, 0.15)
```

It's hard to tell how well this "fits", so we create a QQ plot.

```python
fig, ax = plt.subplots()
res = stats.probplot(hour_sleep, sparams=(l), dist=stats.poisson, plot=ax)
ax.set(title='QQ-Plot: Awakenings, Poisson Distribution', 
       xlabel='Theoretical Quantiles, Poisson Distribution',
       ylabel='Sample Quantiles, Number of Awakenings');
```

This QQ plot looks pretty good!

However, when we represented the fitted Poisson model as a barplot, its shape may have been very familiar. Even though it is a continuous model, let's try a normal model.

```python
plt.hist(awakes, density=True, range=(0,60), bins='auto');
plt.title('Histogram of Number of Awakenings')
plt.xlabel('Awakenings')
plt.ylabel('Density')
plt.grid()
mu, sigma = stats.norm.fit(awakes)

grid = np.linspace(0, 60, 100)
plt.plot(grid, stats.norm.pdf(grid, mu, sigma), label=f'$\hat\mu={mu:.2f}$, $\sigma^2={sigma**2:.2f}$')
```

The fitted normal model on top of the histogram looks pretty good…

```python
fig, ax = plt.subplots()
res = stats.probplot(awakes, sparams=(mu, sigma), dist=stats.norm, plot=ax)
ax.set(title='QQ-Plot: Awakenings, Normal Distribution', 
       xlabel='Theoretical Quantiles, Normal Distribution',
       ylabel='Sample Quantiles, Number of Awakenings');
```

QQ plot looks good as well! Both of these models are reasonable. This is largely because a normal distribution is a decent approximation of a Poisson distribution, especially for "large" values of λ.

## River Length, Gamma Model

We will try to model the length of the rivers, which in this dataset is given in kilometers.

```python
plt.hist(nz_rivers.Length, density=True, range=(0,400), bins=20);
plt.title('NZ River Lengths')
plt.xlabel('River Length (km)')
plt.ylabel('Density')
plt.grid()
```

By looking at the histogram, and simply understanding that rivers will have some positive length, we would like to consider probability distributions that places density on positive numbers. A number of probability distributions have this feature, including: chi-square, F, exponential, and gamma. Let’s try gamma.

Recall that the probability density function of a gamma random variable is given by
$$
f(x|\alpha,\beta)=\frac{1}{\Gamma(\alpha)\beta^\alpha} x^{\alpha-1}e^{-x/\beta},\quad \alpha>0,\ \beta>0
$$

Note that Python will use shape instead of α and scale instead of β.

Also, the gamma distribution has
$$
E[X] = \alpha\beta\\
Var(X)=\alpha\beta^2
$$
Below are plots of gamma with various values for the parameters. By estimating these parameters from the data, hopefully we can find a gamma distribution that matches the shape of the above histogram.
![](img/gamma.png)

While it is not possible to find an analytic solution for the MLE, given $X_1,X_2,\ldots,X_n$ assumed to be a random sample from a gamma distribution, we can use method of moments to obtain estimators:
$$
\tilde\alpha = \frac{\bar{x}^2}{\overline{x^2}-\bar{x}^2}\\
\tilde\beta = \frac{\overline{x^2}-\bar{x}^2}{\bar{x}}
$$

Otherwise, the MLE estimators can be approximated by numerical optimization, as is done by the scipy fit method. Note, we'll maintaint the loc parameter fixed at 0 during the optimization. This is an example of constrained optimization.

```python
plt.hist(nz_rivers.Length, density=True, range=(0,400), bins=20);
plt.title('NZ River Lengths')
plt.xlabel('River Length (km)')
plt.ylabel('Density')
plt.grid()

shape, loc, scale = stats.gamma.fit(nz_rivers.Length, floc=0)

grid = np.linspace(0, 400, 100)
plt.plot(grid, stats.gamma.pdf(grid, shape, loc=loc, scale=scale),
         label=rf'$\alpha={shape:.2f}$, $\beta={scale:.2f}$')
plt.legend()
```

Then to better verify the fit, we create a QQ plot.

```python
fig, ax = plt.subplots()
res = stats.probplot(nz_rivers.Length, sparams=(shape, loc, scale), dist=stats.gamma, plot=ax)
ax.set(title='QQ-Plot: NZ River Length, Gamma Distribution', 
       xlabel='Theoretical Quantiles, Gamma Distribution',
       ylabel='Sample Quantiles, River Length');
```

This QQ plot isn't the best, but it isn't terrible.

## Cocaine Seizure Data

This dataset comes from STRIDE, the System to Retrieve Information from Drug Evidence. It contains all concaine seizures in the US from 2007 that have a known weight.

```python
cocaine = pd.read_csv('../data/cocaine.csv')
cocaine.head()
```

### Cocaine Potency, Beta Model
First, let’s model the potency of the cocaine seized.

```python
cocaine.potency.describe()
```

```python
plt.hist(cocaine.potency, density=True, range=(0,100), bins=20);
plt.title('Potency of Seized Cocaine')
plt.xlabel('Potency  (0% = all filler, 100% = pure cocaine)')
plt.ylabel('Density')
plt.grid()
```

We see that potency can range from 0 (all filler) to 100 (pure cocaine). We aren't aware of any distributions on the interval from 0 to 100, but we do know one on the interval from 0 to 1, the beta distribution.

In order to use the beta distribution, we'll first perform a quick transformation.

```python
coke_pot = cocaine.potency / 100
coke_pot.describe()
```

```python
plt.hist(coke_pot, density=True, range=(0,1), bins=20);
plt.title('Potency of Seized Cocaine')
plt.xlabel('Potency  (0 = all filler, 1 = pure cocaine)')
plt.ylabel('Density')
plt.grid()
```

Recall, that a random variable X that follows a beta distribution has probability density function
$$
f(x|\alpha,\beta) = \left[ \frac{\Gamma(\alpha+\beta)}{\Gamma(\alpha)\Gamma(\beta)} \right]
x^{\alpha-1}(1-x)^{\beta-1},\quad 0\leq x \leq 1,\ \alpha>0,\ \beta>0
$$
Also, the beta distribution has
$$
E[X]=\frac{\alpha}{\alpha+\beta}\\
Var(X)=\frac{\alpha\beta}{(\alpha+\beta)^2(\alpha+\beta+1)}
$$
Below are plots of beta with various values for the parameters. By estimating these parameters from the data, hopefully we can find a gamma distribution that matches the shape of the above histogram.

![](img/beta.png)

While it is not possible to find an analytic solution for the MLE, we can use method of moments to obtain estimators:
$$
\tilde\alpha = \bar{x}\left(\frac{\bar{x}(1-\bar{x})}{s^2}-1\right)\\
\tilde\beta = (1-\bar{x})\left(\frac{\bar{x}(1-\bar{x})}{s^2}-1\right)
$$

We can also approximate the MLE estimators using numerical optimization, as done by scipy.

```python
a, b, loc, scale = stats.beta.fit(coke_pot, floc=0, fscale=1)
```

```python
plt.hist(coke_pot, density=True, range=(0,1), bins=20);
plt.title('Potency of Seized Cocaine')
plt.xlabel('Potency  (0 = all filler, 1 = pure cocaine)')
plt.ylabel('Density')
plt.grid()

grid = np.linspace(0, 1, 100)
plt.plot(grid, stats.beta.pdf(grid, a, b, loc=loc, scale=scale),
         label=rf'$\alpha={a:.2f}$, $\beta={b:.2f}$')
plt.legend()
```

This looks pretty good, but let's verify with a QQ plot.

```python
fig, ax = plt.subplots()
res = stats.probplot(coke_pot, sparams=(a, b, loc, scale), dist=stats.beta, plot=ax)
ax.set(title='QQ-Plot: Cocaine Potency, Beta Distribution', 
       xlabel='Theoretical Quantiles, Beta Distribution',
       ylabel='Sample Quantiles, Cocaine Potency');
```

The left tail isn't perfect, but again, this look pretty good.

### Cocaine Price, Some Model?
Now let's try to model the price of the cocaine seized.

```python
plt.hist(cocaine.price, density=True, bins=18);
plt.title('Price of Seized Cocaine')
plt.xlabel('Estimate Value, USD')
plt.ylabel('Density')
plt.grid()
```

This is a very right-skewed distribution. We know a number of distributions that are valide for positive numbers. Let's try exponential.
$$
f(x|\lambda) = \lambda e^{-\lambda x},\quad x>0,\ \lambda>0
$$

Note that the MLE for $\lambda$ is $1/\bar{X}$.

```python
# scale corresponds to 1/lambda
loc, scale = stats.expon.fit(cocaine.price, floc=0)
```

```python
plt.hist(cocaine.price, density=True, bins=18);
plt.title('Price of Seized Cocaine')
plt.xlabel('Estimate Value, USD')
plt.ylabel('Density')
plt.grid()

grid = np.linspace(0, 9000, 100)
plt.plot(grid, stats.expon.pdf(grid, loc, scale),
         label=rf'$1/\lambda={scale:.2f}$')
plt.legend()
```

This looks OK, but let's check with a QQ plot.

```python
fig, ax = plt.subplots()
res = stats.probplot(cocaine.price, sparams=(loc, scale), dist=stats.expon, plot=ax)
ax.set(title='QQ-Plot: Cocaine Price, Exponential Distribution', 
       xlabel='Theoretical Quantiles, Exponential Distribution',
       ylabel='Sample Quantiles, Cocaine Price');
```

This actually highlights an issue. It seems that this distribution actually fits poorly in the right tail, which is heavier than the exponential distribution.

We could try some other distributions, but let's cheat a bit a use something called a kernel density estimate.

```python
kernel = stats.gaussian_kde(cocaine.price)
```

```python
plt.hist(cocaine.price, density=True, bins=18);
plt.title('Price of Seized Cocaine')
plt.xlabel('Estimate Value, USD')
plt.ylabel('Density')
plt.grid()

grid = np.linspace(0, 9000, 100)
plt.plot(grid, kernel(grid))
```

Wow, look at that, it fits great, which we can verify with a QQ plot.
